**1. Devise an experiment to verify that the list index operator is ùëÇ(1)**

```python
from timeit import Timer

list_index = Timer("x[10]", "from __main__ import x")

for i in range(1_000_000, 100_000_001, 1_000_000):
    x = list(range(i))
    print(f"x[10]: {list_index.timeit(number=1000):10.5f} milliseconds")
```

**2. Devise an experiment to verify that get item and set item are ùëÇ(1) for dictionaries.**

```python
import timeit
import random

print(f"{'n':10s}{'get':>10s}{'set':>10s}")
for i in range(10_000, 1_000_001, 50_000):
    get_t = timeit.Timer(
        f"x[random.randrange({i})] = None",
        "from __main__ import random, x"
    )
    set_t = timeit.Timer(
        f"x.get(random.randrange({i}))",
        "from __main__ import random, x"
        )
    x = {j: None for j in range(i)}
    get_time = get_t.timeit(number=1000)
    x = {j: None for j in range(i)}
    set_time = set_t.timeit(number=1000)
    print(f"{i:<10,}{get_time:>10.3f}{set_time:>10.3f}")
```

**3. Devise an experiment that compares the performance of the del operator on lists and dictionaries.**

```python
import timeit
import random

def del_dict_items(x):
    random_index = random.randrange(len(x) - 1)
    try:
        del x[random_index]
    except KeyError:
        x.setdefault(random_index, None)
        del x[random_index]

print(f"{'n':10s}{'list':>10s}{'dict':>10s}")
for i in range(10000, 100001, 1000):
    t_list = timeit.Timer("del x[random.randrange(len(x)-1)]", "from __main__ import random, x")
    t_dict = timeit.Timer("del_dict_items(x)", "from __main__ import random, x, del_dict_items")
    x = list(range(i))
    list_del_time = t_list.timeit(number=1000)
    x = {j:None for j in range(i)}
    dict_del_time = t_dict.timeit(number=1000)
    print(f"{i:<10,}{list_del_time:>10.3f}{dict_del_time:>10.3f}")
```

**4. Given a list of numbers in random order, write an algorithm that works in ùëÇ(ùëõlog(ùëõ)) to find the kth smallest number in the list.**

```python
import random

my_list = [random.randrange(10, 30) for _ in range(10)]
my_list
```

```python
def find_kth_log(num_list, k):
    return sorted(set(num_list))[k-1]

find_kth_log(my_list, 2)
```
